```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(filename='img2.jpg')
img = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2RGB)
plt.imshow(X=img);
```

```python
img[0:1]
```

```python
img[0:1].size
```

## Resmi gri tonlamaya dönüştürelim

```python
gray_img = cv2.cvtColor(src=img, code=cv2.COLOR_RGB2GRAY)
plt.imshow(X=gray_img, cmap='gray');
```

## Resmi yeniden boyutlandıralım

```python
img.shape
```

```python
resized_img = cv2.resize(src=img, dsize=(1100, 800))
plt.imshow(X=resized_img);
```

## Görselde kenarları tespit edelim.

```python
canny = cv2.Canny(image=img, threshold1=120, threshold2=200)
plt.imshow(X=canny);
```

```python
canny2 = cv2.Canny(image=img, threshold1=40, threshold2=120)
plt.imshow(X=canny2);
```

```python
kernel = np.ones((5, 5), dtype=np.uint8)
gradient = cv2.morphologyEx(src=img, op=cv2.MORPH_GRADIENT, kernel=kernel)
plt.imshow(X=gradient);
```

```python
laplacian = cv2.Laplacian(src=img, ddepth=cv2.CV_64F, ksize=3)
plt.imshow(X=laplacian);
```

## Histogram eşitleme işlemi yapalım.

```python
hist = cv2.calcHist(images=img, channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist);
```

```python
eq_img = cv2.cvtColor(src=img, code=cv2.COLOR_RGB2GRAY)
equalizeHist = cv2.equalizeHist(src=eq_img)
plt.imshow(X=equalizeHist, cmap='gray');
```

```python
plt.imshow(X=eq_img, cmap='gray');
```

## Görseli Gauss filtresi ile bulanıklaştıralım.

```python
gauss = cv2.GaussianBlur(src=img, ksize=(5, 5), sigmaX=0.5, sigmaY=0.5)
plt.imshow(X=gauss);
```

## Görselin parlaklığı ile oynayalım.

```python
scale = cv2.convertScaleAbs(src=img, alpha=0.5, beta=0.5)
plt.imshow(X=scale);
```

```python
scale2 = cv2.convertScaleAbs(src=img, alpha=0.8, beta=0.2)
plt.imshow(X=scale2);
```

```python
scale3 = cv2.convertScaleAbs(src=img, alpha=0.2, beta=0.8)
plt.imshow(X=scale3);
```

## Görsel üzerinde maskeleme işlemi yapalım.

```python
mask = np.zeros(img.shape[:2], dtype=np.uint8)
plt.imshow(X=mask, cmap='gray');
```

```python
mask[500:700, 450:820] = 255
plt.imshow(X=mask, cmap='gray');
```

```python
clip = cv2.bitwise_and(src1=img, src2=img, mask=mask)
plt.imshow(X=clip);
```
